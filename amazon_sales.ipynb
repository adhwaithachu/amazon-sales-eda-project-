```python
#Question1: What is the average rating of products in each category?

import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("amazon.csv")
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')

avg_rating = df.groupby('category')['rating'].mean().sort_values(ascending=False).head(20)

avg_rating.plot(kind='bar', color='red', figsize=(10,5))
plt.title("Top 20 Categories by Average Rating")
plt.xlabel("Category")
plt.ylabel("Average Rating")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
#Question2: What is the most common rating given by customers?

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("amazon.csv")
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')

sns.countplot(x='rating', data=df)
plt.title("Most Common Ratings Given by Customers")
plt.xlabel("Rating")
plt.ylabel("Count")
plt.show()
```

```python
#Question3: How many products have ratings above 4.0?
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("amazon.csv")
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')

df['rating_above_4'] = df['rating'] > 4.0

sns.countplot(x='rating_above_4', data=df)
plt.title("Number of Products with Ratings Above 4.0")
plt.xlabel("Rating > 4.0")
plt.ylabel("Count")
plt.show()
```

```python
#Question4:How do ratings vary across categories?

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("amazon.csv").head(50)

df['rating'] = pd.to_numeric(df['rating'], errors='coerce')

plt.figure(figsize=(10,6))
sns.boxplot(x='category', y='rating', data=df)
plt.title("Ratings Across Different Categories (Top 50 Products)")
plt.xlabel("Category")
plt.ylabel("Rating")
plt.xticks(rotation=45)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("amazon.csv")

df['rating_count'] = pd.to_numeric(df['rating_count'], errors='coerce')

top_products = df.nlargest(10, 'rating_count')

plt.bar(top_products['product_name'], top_products['rating_count'], color='green')
plt.title("Top 10 Products with Highest Rating Count")
plt.xlabel("Product Name")
plt.ylabel("Rating Count")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```
